# 1. Basic Count

```python
# Filter the dataset to include only rock-type Pokémon
rock_pokemon = df[(df['type1'] == 'rock') | (df['type2'] == 'rock')].reset_index(drop=True)
print(f"Total Rock-type Pokémon: {rock_pokemon.shape[0]}")
```

# 2. Power Analysis (usually the 'Total' or 'Attack' column)

```python
# Plot distribution of Total power
plt.figure(figsize=(10, 5))
sns.histplot(df_rock['base_total'], kde=True, bins=15, color='orange')
plt.title('Distribution of Total Power - Rock Type')
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Boxplot for power
plt.figure(figsize=(6, 4))
sns.boxplot(x=df_rock['base_total'], color='lightblue')
plt.title('Boxplot of Total Power - Rock Type')
plt.xlabel('Total')
plt.show()

# Most and least powerful Pokémon
most_powerful = df_rock.loc[df_rock['base_total'].idxmax()]
least_powerful = df_rock.loc[df_rock['base_total'].idxmin()]
print(f"Most powerful Rock Pokémon: {most_powerful['name']} with Total = {most_powerful['base_total']}")
print(f"Weakest Rock Pokémon: {least_powerful['name']} with Total = {least_powerful['base_total']}")
```

# 3. Defense Analysis

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_rock['defense'], kde=True, bins=15, color='green')
plt.title('Distribution of Defense - Rock Type')
plt.xlabel('Defense')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x=df_rock['defense'], color='lightgreen')
plt.title('Boxplot of Defense - Rock Type')
plt.xlabel('Defense')
plt.show()

# Highest and lowest defense
strongest_defense = df_rock.loc[df_rock['defense'].idxmax()]
weakest_defense = df_rock.loc[df_rock['defense'].idxmin()]
print(f"Highest Defense Rock Pokémon: {strongest_defense['name']} with {strongest_defense['defense']}")
print(f"Lowest Defense Rock Pokémon: {weakest_defense['name']} with {weakest_defense['defense']}")
```

# 4. Gender Distribution (if you have a 'Percentage Male' column)

```python
# Some Pokémon may have NaN in 'percentage_male', remove them
gender_data = df_rock.dropna(subset=['percentage_male'])

avg_male_percentage = gender_data['percentage_male'].mean()
avg_female_percentage = 100 - avg_male_percentage

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie([avg_male_percentage, avg_female_percentage], labels=['Male', 'Female'], autopct='%1.1f%%', colors=['blue', 'pink'])
plt.title('Average Gender Distribution - Rock Type')
plt.show()

print(f"Average Male Percentage: {avg_male_percentage:.2f}%")
```

# 5. Legendary Pokémon

```python
legendary_count = df_rock[df_rock['is_legendary'] == True].shape[0]
print(f"Number of Legendary Rock-type Pokémon: {legendary_count}")
```

# 6. Relationship to other types (what types are weak/strong against)

```python
# Select columns that start with 'against_'
against_cols = [col for col in df.columns if col.startswith('against_')]

# Average damage received by Rock type Pokémon
avg_damage_received = df_rock[against_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_damage_received.plot(kind='bar', color='coral')
plt.title('Average Damage Received by Rock Type from Other Types')
plt.ylabel('Damage Multiplier')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Most and least damage received
most_dangerous = avg_damage_received.idxmax().replace('against_', '').capitalize()
least_dangerous = avg_damage_received.idxmin().replace('against_', '').capitalize()

print(f"Rock-type Pokémon receive most damage from: {most_dangerous}")
print(f"Rock-type Pokémon receive least damage from: {least_dangerous}")

# Average damage dealt *by* Rock-type to other types
# Use all Rock-type Pokémon and average their 'against_X' columns
avg_damage_by_rock = df.loc[(df['type1'] == 'rock') | (df['type2'] == 'rock'), against_cols].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
avg_damage_by_rock.plot(kind='bar', color='slateblue')
plt.title('Average Damage Dealt by Rock Type to Other Types')
plt.ylabel('Damage Multiplier')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

best_against = avg_damage_by_rock.idxmax().replace('against_', '').capitalize()
worst_against = avg_damage_by_rock.idxmin().replace('against_', '').capitalize()

print(f"Rock-type Pokémon are strongest against: {best_against}")
print(f"Rock-type Pokémon are weakest against: {worst_against}")
```

# 7. Speed, Height, Weight

```python
for column in ['speed', 'height_m', 'weight_kg']:
    plt.figure(figsize=(10, 5))
    sns.histplot(df_rock[column], kde=True, bins=15)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_rock[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()

    mean_val = df_rock[column].mean()
    median_val = df_rock[column].median()
    min_val = df_rock.loc[df_rock[column].idxmin()]
    max_val = df_rock.loc[df_rock[column].idxmax()]

    print(f"{column.capitalize()} - Mean: {mean_val:.2f}, Median: {median_val:.2f}")
    print(f"Min {column}: {min_val['name']} ({min_val[column]})")
    print(f"Max {column}: {max_val['name']} ({max_val[column]})")
```

# 8. Strategic Choice

```python
# Let's consider Base Total, Defense, Speed, and Legendary status
strategic_candidates = df_rock.sort_values(by=['base_total', 'defense', 'speed'], ascending=False)

# Select top 3 candidates
strategic_top3 = strategic_candidates.head(3)[['name', 'base_total', 'defense', 'speed', 'is_legendary']]
print("Top strategic Rock-type Pokémon candidates:\n")
print(strategic_top3)

# Choose one
chosen = strategic_top3.iloc[0]
print(f"\nChosen Pokémon for the tournament: {chosen['name']}")
print("Reason: High total stats, good defense and speed, and legendary status" if chosen['is_legendary'] else "Reason: Strong overall performance and balance.")
```
