## Ejercicio

Graficar sin(x) y cos(x) en la misma figura

```python
x = np.linspace(0, 2*np.pi, 5)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(4, 4))
plt.xlabel('x')
plt.xticks(np.linspace(0, 2*np.pi, 10))
plt.xticks(rotation=45)

plt.ylabel('y')
plt.title('Funciones')
plt.grid()
plt.plot(x, y1, label='y=sin(x)', color='green', linewidth=3, linestyle='--')
plt.plot(x, y2, label='y=cos(x)')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

t = np.linspace(0, 4*np.pi, 100)
x = np.cos(t)
y = np.sin(t)
z = t

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Funciones')

ax.plot(x, y, z)

plt.show()
```

### Ejercicio 1: Gráfico de Barras

Crea un gráfico de barras que muestre las ventas mensuales (enero a junio) de una tienda.

Usa los datos:
+ ventas = [120, 85, 90, 150, 200, 95].

Añade etiquetas a las barras y coloréalas de verde.

```python
# Importar librerías necesarias
import matplotlib.pyplot as plt

# 1. Preparación de datos
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio']
ventas = [120, 85, 90, 150, 200, 95]
colores_verdes = ['green']*6
#colores = ['#2ecc71', '#27ae60', '#16a085', '#1abc9c', '#3498db', '#2980b9']

# 2. Creación de la figura y ejes
fig, ax = plt.subplots(figsize=(10, 6))

# 3. Creación del gráfico de barras con personalización
barras = ax.bar(meses, ventas, color=colores_verdes, edgecolor='black', linewidth=1)

"""
barras = ax.bar(meses, 
    ventas, 
    color=colores,          # Colores personalizados para cada barra
    edgecolor='black',      # Borde negro para las barras
    linewidth=1            # Grosor del borde
)
"""

# 4. Personalización adicional
# Añadir etiquetas a las barras
ax.bar_label(barras, labels=ventas, padding=3)

# Rotar etiquetas del eje X
plt.xticks(rotation=45)

# Añadir título y etiquetas
ax.set_title('Ventas Mensuales (Enero - Junio)', fontsize=15)
ax.set_xlabel('Meses', fontsize=12)
ax.set_ylabel('Ventas (unidades)', fontsize=12)

# Añadir cuadrícula
ax.grid(axis='y', linestyle='--', alpha=0.7)

# 5. Mostrar el gráfico
plt.tight_layout()  # Ajuste automático del layout
plt.show()
```

### Ejercicio 2: Gráfico de pastel (Pie Chart)

Visualiza el porcentaje de uso de navegadores web por parte de los usuarios.

Datos:
+ navegadores = ['Chrome', 'Firefox', 'Edge', 'Safari', 'Otros']
+ porcentajes = [65, 15, 10, 7, 3]

Requisitos:
+ Muestra los porcentajes en cada porción.
+ Explota la porción de “Chrome”.
+ Título: "Uso de Navegadores Web".

```python
import matplotlib.pyplot as plt

# Datos proporcionados
navegadores = ['Chrome', 'Firefox', 'Edge', 'Safari', 'Otros']
porcentajes = [65, 15, 10, 7, 3]
colores = ['#ffff00', '#FF9500', '#4285F4', '#2271b3', '#AAAAAA']  # Colores representativos de naveagadores

# Configurar la "explosión" - Chrome se destacará
explode = [0.1, 0, 0, 0, 0]  # Solo Chrome se separa (10% del radio)

# Crear la figura
plt.figure(figsize=(8, 8))

# Crear el gráfico de pastel
plt.pie(
    porcentajes,
    labels=navegadores,
    autopct='%1.1f%%',  # Muestra porcentajes con 1 decimal
    startangle=90,     # Ángulo inicial (vertical arriba)
    colors=colores,
    explode=explode,   # Destacar Chrome
    shadow=True,       # Sombra para efecto 3D
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}  # Bordes blancos
)

# Añadir título
plt.title('Distribución de Navegadores Web', fontsize=16, pad=2)
plt.legend(navegadores, title="Navegadores", loc="best")

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

### Ejercicio 3

Selecciona la gráfica que consideres más conveniente para observar y comparar el rendimiento de tres estudiantes (Juan, Ana y Luis) durante 4 parciales.

Datos:
+ parciales = ['Parcial 1', 'Parcial 2', 'Parcial 3', 'Parcial 4']
+ juan = [3.0, 3.5, 4.0, 4.2]
+ ana = [4.5, 4.6, 4.7, 4.9]
+ luis = [2.8, 3.1, 3.3, 3.5]

Requisitos:
+ Agregar leyenda con los nombres.
+ Título: "Evolución del Rendimiento Académico".

```python
import matplotlib.pyplot as plt

# Datos proporcionados
parciales = ['Parcial 1', 'Parcial 2', 'Parcial 3', 'Parcial 4']
juan = [3.0, 3.5, 4.0, 4.2]
ana = [4.5, 4.6, 4.7, 4.9]
luis = [2.8, 3.1, 3.3, 3.5]

# Configurar el gráfico
plt.figure(figsize=(10, 6))  # Tamaño del gráfico

# Crear las líneas para cada estudiante
plt.plot(parciales, juan, marker='o', label='Juan', linewidth=2, color='#1f77b4')  # Azul
plt.plot(parciales, ana, marker='s', label='Ana', linewidth=2, color='#ff7f0e')   # Naranja
plt.plot(parciales, luis, marker='^', label='Luis', linewidth=2, color='#2ca02c')  # Verde

# Personalización del gráfico
plt.title('Evolución del Rendimiento Académico', fontsize=16, pad=20)
plt.xlabel('Parciales', fontsize=12)
plt.ylabel('Calificación', fontsize=12)
plt.legend(title='Estudiantes', title_fontsize='13', fontsize=12)  # Leyenda con título
plt.grid(True, linestyle='--', alpha=0.7)  # Cuadrícula

# Ajustar límites del eje Y para mejor visualización
plt.ylim(2.5, 5.0)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python

import numpy as np

x = np.arange(len(parciales))  # Posiciones de los parciales
width = 0.25  # Ancho de las barras

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, juan, width, label='Juan')
rects2 = ax.bar(x, ana, width, label='Ana')
rects3 = ax.bar(x + width, luis, width, label='Luis')

ax.set_title('Comparación de Rendimiento por Parcial')
ax.set_xticks(x)
ax.set_xticklabels(parciales)
ax.legend()
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)

plt.show()
```

### Ejercicio 4: Subplots

Creee una figura con 2 subplots:
+ Gráfico de barras de ventas semanales.
+ Gráfico de línea de ganancias semanales.

Datos:
+ dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
+ ventas = [200, 220, 180, 240, 260]
+ ganancias = [50, 60, 45, 70, 80]

Requisitos:
+ Organizar los subplots uno al lado del otro (1 fila x 2 columnas).
+ Incluir títulos para cada subplot.
+ Etiquetas de ejes y colores diferentes.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
dias = ['Lun', 'Mar', 'Mié', 'Jue', 'Vie']
ventas = [200, 220, 180, 240, 260]
ganancias = [50, 60, 45, 70, 80]

# Crear figura con 2 subplots (1 fila, 2 columnas)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))  # 12" ancho, 5" alto

## Primer subplot: Gráfico de barras de ventas
ax1.bar(dias, ventas, color='#1f77b4', edgecolor='black')  # Azul con bordes negros
ax1.set_title('Ventas Semanales', fontsize=14)
ax1.set_xlabel('Días de la semana', fontsize=12)
ax1.set_ylabel('Cantidad vendida', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir etiquetas a las barras
for bar in ax1.patches:
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 5,
                f'{int(bar.get_height())}',
                ha='center', va='bottom')

## Segundo subplot: Gráfico de línea de ganancias
ax2.plot(dias, ganancias, marker='o', color='#ff7f0e', linewidth=2, markersize=8)  # Naranja
ax2.set_title('Ganancias Semanales', fontsize=14)
ax2.set_xlabel('Días de la semana', fontsize=12)
ax2.set_ylabel('Ganancias ($)', fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.7)

# Añadir etiquetas a los puntos
for i, txt in enumerate(ganancias):
    ax2.annotate(f'${txt}', (dias[i], ganancias[i]),
                    textcoords="offset points", xytext=(0,10),
                    ha='center')

# Ajustar espacio entre subplots
plt.tight_layout(pad=3.0)

# Mostrar la figura
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import StrMethodFormatter
import matplotlib.patheffects as path_effects

# Configuración global de estilo manual (sin depender de seaborn)
plt.rcParams.update({
    'font.family': 'DejaVu Sans',  # O usa 'sans-serif'
    'axes.facecolor': '#f8f9fa',
    'axes.edgecolor': '#333333',
    'axes.grid': True,
    'grid.alpha': 0.3,
    'grid.color': '#333333',
    'axes.labelcolor': '#333333',
    'axes.titlesize': 14,
    'axes.titleweight': 'bold',
    'axes.labelsize': 12,
    'xtick.color': '#333333',
    'ytick.color': '#333333',
    'figure.facecolor': 'white',
    'figure.titlesize': 16
})

# Datos
dias = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes']
ventas = [200, 220, 180, 240, 260]
ganancias = [50, 60, 45, 70, 80]

# Crear figura con especificaciones avanzadas
fig = plt.figure(figsize=(14, 6), dpi=100)
gs = fig.add_gridspec(1, 2, wspace=0.3, hspace=0.4)
ax1 = fig.add_subplot(gs[0])
ax2 = fig.add_subplot(gs[1])

## ------------------
## SUBPLOT 1: VENTAS
## ------------------
# Configuración de barras con efectos
bar_width = 0.6
edge_color = '#1a2e45'
fill_colors = ['#4285f4', '#5a95f5', '#72a4f7', '#8ab4f8', '#a2c3fa']

bars = ax1.bar(dias, ventas, width=bar_width, linewidth=1.5,
                edgecolor=edge_color, color=fill_colors, zorder=3)

# Efecto de profundidad para barras
for bar in bars:
    bar.set_path_effects([
        path_effects.SimpleLineShadow(alpha=0.2),
        path_effects.Normal()
    ])

# Etiquetas inteligentes
for i, bar in enumerate(bars):
    height = bar.get_height()
    va = 'bottom' if height < 250 else 'top'
    offset = 5 if height < 250 else -5
    color = '#333333' if height < 250 else 'white'
    ax1.text(bar.get_x() + bar.get_width()/2, height + offset,
                f'{height}',
                ha='center', va=va, color=color, fontsize=10,
                bbox=dict(facecolor='white' if height < 250 else '#1a2e45',
                            alpha=0.8, edgecolor='none', boxstyle='round'))

# Personalización de ejes
ax1.set_title('ANÁLISIS DE VENTAS SEMANALES', pad=20)
ax1.set_xlabel('Días de la Semana', labelpad=10)
ax1.set_ylabel('Unidades Vendidas', labelpad=10)
ax1.yaxis.set_major_formatter(StrMethodFormatter('{x:,.0f}'))
ax1.set_ylim(0, 300)

# Línea de promedio
avg_ventas = np.mean(ventas)
ax1.axhline(avg_ventas, color='#ea4335', linestyle='--', linewidth=1.5)
ax1.text(len(dias)-0.5, avg_ventas+10, f'Promedio: {avg_ventas:.0f}',
            color='#ea4335', ha='right', va='center')

## -------------------
## SUBPLOT 2: GANANCIAS
## -------------------
# Gráfico de línea con marcadores personalizados
line, = ax2.plot(dias, ganancias, color='#fbbc05', linewidth=3,
                    marker='o', markersize=10, markerfacecolor='white',
                    markeredgewidth=2, zorder=3)

# Área sombreada
ax2.fill_between(dias, ganancias, color='#fbbc05', alpha=0.1)

# Etiquetas de datos
for i, (dia, gan) in enumerate(zip(dias, ganancias)):
    offset = 15 if gan < 70 else -15
    arrowprops = dict(arrowstyle='-|>', color='#333333', linewidth=1)
    ax2.annotate(f'${gan:.0f}',
                    xy=(dia, gan),
                    xytext=(0, offset),
                    textcoords='offset points',
                    ha='center', va='center',
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
                    arrowprops=arrowprops)

# Personalización de ejes
ax2.set_title('EVOLUCIÓN DE GANANCIAS', pad=20)
ax2.set_xlabel('Días de la Semana', labelpad=10)
ax2.set_ylabel('Ganancias (USD)', labelpad=10)
ax2.yaxis.set_major_formatter(StrMethodFormatter('${x:,.0f}'))
ax2.set_ylim(0, 100)

# Línea de tendencia
z = np.polyfit(range(len(ganancias)), ganancias, 1)
p = np.poly1d(z)
ax2.plot(dias, p(range(len(ganancias))), '--', color='#34a853', linewidth=2)
ax2.text(len(dias)-0.5, p(len(ganancias)-1)+5, f'Tendencia: +${p[1]:.1f}/día',
            color='#34a853', ha='right')

## -------------------
## AJUSTES FINALES
## -------------------
# Título principal
fig.suptitle('REPORTE SEMANAL - TIENDA XYZ', y=0.98, fontsize=16, fontweight='bold')

# Cuadro de información
info_text = f"""Datos correspondientes a la semana del 10-14 Junio 2024
• Mejor día de ventas: {dias[np.argmax(ventas)]} ({max(ventas)} unidades)
• Mayor ganancia: {dias[np.argmax(ganancias)]} (${max(ganancias)})
"""
fig.text(0.5, 0.02, info_text, ha='center', fontsize=10,
            bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

plt.tight_layout()
plt.show()
```
